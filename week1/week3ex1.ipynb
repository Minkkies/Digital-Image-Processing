```python
import numpy as np
import cv2

img = cv2.imread('./week_introduction/lena.png') 
```

```python
print(f"img type: {type(img)}")  # <class 'numpy.ndarray'>
print(f"img shape: {img.shape}")  # (512, 512, 3)
```

```python
def show_image(img):
    cv2.imshow('Lena Image', img)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
    for i in range(1,5): # แสดงภาพ 4 ครั้ง โดยมีดีเลย์ 1 วินาที
        cv2.waitKey(1000)  # wait for 1 second
```

```python
show_image(img)
```

```python
img = img.astype(np.float16)

gray_img = (0.1140 * img[:, :, 0] + 0.5870 * img[:, :, 1] + 0.2989 * img[:, :, 2]).astype(np.uint8)
show_image(gray_img)
```
