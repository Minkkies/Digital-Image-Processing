# วิธีนำรูปภาพไปใช้ อ่านค่าสีจากภาพ แบบสีกับ GRAYSCALE

```python
import numpy as np
import cv2 #สำหรับการอ่าน/ประมวลผลภาพ
#เก็บภาพเป็น array 3ชั้น RGB 
#จากนั้นแปลงเป็น PIL Image เพื่อใช้งานฟังก์ชันต่างๆของ PIL
```

```python

img1 = cv2.imread('./week_introduction/lena.png',cv2.IMREAD_COLOR)#คืนภาพ BGR 3 ช่อง (shape เป็น (height, width, 3)
img2 = cv2.imread('./week_introduction/lena.png',cv2.IMREAD_GRAYSCALE)#คืนภาพ 2D (shape เป็น (height, width)

# ตรวจสอบก่อนว่า img1 และ img2 ไม่ใช่ None
if img1 is not None and img2 is not None:
    print("Image 1 Shape:", img1.shape)
    print("Image 2 Shape:", img2.shape)
elif img1 is not None:
    print("Image 1 Shape:", img1.shape)
    print("Image 2 โหลดไม่สำเร็จ")
elif img2 is not None:
    print("Image 1 โหลดไม่สำเร็จ")
    print("Image 2 Shape:", img2.shape)
else:
    print("ไม่สามารถแสดง Shape ได้: ภาพทั้งสองยังไม่ถูกโหลดสำเร็จ")
```
