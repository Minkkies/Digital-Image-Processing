# Grayscale images
- ‡∏à‡∏∞‡πÄ‡∏´‡∏•‡∏∑‡∏≠‡πÄ‡∏û‡∏µ‡∏¢‡∏á 3 -> 1 byte (1 space) ‡πÄ‡∏û‡∏£‡∏≤‡∏∞‡πÄ‡∏≠‡∏≤‡∏ó‡∏±‡πâ‡∏á 3 ‡∏°‡∏≤‡∏ï‡πà‡∏≠‡∏Å‡∏±‡∏ô‡πÅ‡∏ö‡∏ö linear
- ‡πÄ‡∏£‡∏µ‡∏¢‡∏Å‡∏•‡∏≥‡∏î‡∏±‡∏ö‡∏Ñ‡πà‡∏≤‡∏™‡∏µ‡∏à‡∏≤‡∏Å‡∏ô‡πâ‡∏≠‡∏¢ -> ‡∏°‡∏≤‡∏Å ```G -> R -> B```
- ‡∏•‡∏≥‡∏î‡∏±‡∏ö‡∏Å‡∏≤‡∏£‡πÄ‡∏£‡∏µ‡∏¢‡∏áspace ``` B -> G -> R ```
- 3 ‡∏ä‡∏±‡πâ‡∏ô‡∏Ñ‡∏∑‡∏≠ 3‡∏ä‡πà‡∏≠‡∏á‡∏™‡∏µ Blue Green Red ‡πÅ‡∏•‡∏∞‡πÅ‡∏ï‡πà‡∏•‡∏∞‡∏ä‡∏±‡πâ‡∏ô‡πÄ‡∏Å‡πá‡∏ö‡∏Ñ‡∏ß‡∏≤‡∏°‡πÄ‡∏Ç‡πâ‡∏°‡∏™‡∏µ 0-255

### ‡πÇ‡∏Ñ‡πâ‡∏î‡πÅ‡∏™‡∏î‡∏á‡∏£‡∏π‡∏õ‡∏†‡∏≤‡∏û ‡∏î‡πâ‡∏ß‡∏¢‡∏ä‡∏±‡πâ‡∏ô‡∏™‡πÄ‡∏õ‡∏ã B(0) -> G(1) -> R(2)

````python
import cv2
import numpy as np
````

````python
def show_image(img):
    cv2.imshow('Lena Image', img)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
````

````python
img = cv2.imread('./week_introduction/lena.png')
````

````python
show_image(img[:,:])#[:,:] show full image every row and column default is color image
````

````python
#imread ‡∏°‡∏µ 2 args (path, flag(default=1))
# 1 = color, 0 = grayscale, -1 = unchanged

print(cv2.IMREAD_COLOR, "color")
print(cv2.IMREAD_GRAYSCALE, "grayscale")
print(cv2.IMREAD_UNCHANGED, "unchanged")
````

‡∏à‡∏∞‡πÑ‡∏î‡πâ‡∏ú‡∏•‡∏•‡∏±‡∏û‡∏ó‡πå‡∏Å‡∏≤‡∏£‡∏Å‡∏£‡∏≠‡∏á‡πÅ‡∏ï‡πà‡∏•‡∏∞‡∏ä‡∏±‡πâ‡∏ô (img[:,:,0]) ‡∏à‡∏∞‡∏Å‡∏£‡∏≠‡∏á‡∏ä‡∏±‡πâ‡∏ô‡∏ó‡∏µ‡πà‡∏°‡∏µ‡∏™‡∏µ‡∏Ç‡∏≠‡∏á B ‡∏ú‡∏™‡∏°‡∏ñ‡πâ‡∏≤‡∏°‡∏µ‡πÄ‡∏õ‡πá‡∏ô‡∏™‡∏µ‡∏Ç‡∏≤‡∏ß=255 ‡∏™‡∏µ‡∏î‡∏≥=0 
‡∏™‡∏µ‡πÄ‡∏ó‡∏≤=128

### ‡∏™‡πà‡∏ß‡∏ô‡∏≠‡∏±‡∏ô‡∏ô‡∏µ‡πâ‡πÄ‡∏õ‡πá‡∏ô‡∏Å‡∏≤‡∏£‡∏ï‡πà‡∏≠‡∏£‡∏π‡∏õ‡∏ó‡∏±‡πâ‡∏á3‡∏ä‡∏±‡πâ‡∏ô‡πÅ‡∏ö‡∏ö‡πÅ‡∏ô‡∏ß‡∏Ç‡∏ß‡∏≤‡∏á

````python
#‡∏õ‡∏Å‡∏ï‡∏¥ img3.shape ‡∏à‡∏∞‡πÑ‡∏î‡πâ‡πÄ‡∏õ‡πá‡∏ô tuple
#‡πÅ‡∏ï‡πà‡∏û‡∏≠‡πÄ‡∏Ç‡∏µ‡∏¢‡∏ô‡πÅ‡∏ö‡∏ö‡∏ô‡∏µ‡πà‡∏à‡∏∞‡πÑ‡∏î‡πâ int ‡∏ò‡∏£‡∏£‡∏°‡∏î‡∏≤
print(type(img), img.shape)
b, g, r = cv2.split(img) # split color image to 3 channel
print(type(b), b.shape)
print(type(g), g.shape)
print(type(r), r.shape)
````

````python
img2 = cv2.hconcat([b, g, r])  # concatenate 3 channel images side by side 
#row need to be same, column can be different
````

````python
# vconcat = ‡∏ï‡πà‡∏≠‡∏Å‡∏±‡∏ô‡πÅ‡∏ô‡∏ß‡∏ï‡∏±‡πâ‡∏á
img3 = cv2.vconcat([b, g, r])
show_image(img3)
````

````python
show_image(img2)
````

### ‡πÅ‡∏™‡∏î‡∏á‡∏Å‡∏≤‡∏£‡∏£‡∏ß‡∏°‡∏Å‡∏±‡∏ô‡∏Ç‡∏≠‡∏á space

```0.1140*img[i, j, 0] + 0.5870*img[i, j, 1] + 0.2989*img[i, j, 2]```

**‡∏Ñ‡πà‡∏≤‡∏™‡∏±‡∏°‡∏õ‡∏£‡∏∞‡∏™‡∏¥‡∏ó‡∏û‡∏ß‡∏Å‡∏ô‡∏µ‡πâ‡∏Ñ‡∏∑‡∏≠** -> ‡∏°‡∏≤‡∏ï‡∏£‡∏ê‡∏≤‡∏ô‡∏™‡∏≤‡∏Å‡∏•‡∏Ç‡∏≠‡∏á‡∏Å‡∏≤‡∏£‡πÅ‡∏õ‡∏•‡∏á‡∏†‡∏≤‡∏û‡∏™‡∏µ‡πÑ‡∏õ‡πÄ‡∏õ‡πá‡∏ô Grayscale ‡∏ó‡∏µ‡πà‡πÄ‡∏£‡∏µ‡∏¢‡∏Å‡∏ß‡πà‡∏≤:

üìå ITU-R BT.601 (‡∏´‡∏£‡∏∑‡∏≠ Rec.601)

‡πÉ‡∏ä‡πâ‡πÉ‡∏ô‡∏£‡∏∞‡∏ö‡∏ö‡∏ß‡∏¥‡∏î‡∏µ‡πÇ‡∏≠‡πÅ‡∏•‡∏∞‡∏†‡∏≤‡∏û‡∏î‡∏¥‡∏à‡∏¥‡∏ó‡∏±‡∏•‡∏°‡∏≤‡∏ô‡∏≤‡∏ô ‡πÄ‡∏ä‡πà‡∏ô SDTV
‡∏™‡∏π‡∏ï‡∏£‡πÄ‡∏î‡∏¥‡∏°‡∏™‡∏≥‡∏´‡∏£‡∏±‡∏ö‡πÅ‡∏õ‡∏•‡∏á RGB ‚Üí Luminance (‡∏Ñ‡∏ß‡∏≤‡∏°‡∏™‡∏ß‡πà‡∏≤‡∏á)

gray = 0.1140 B + 0.5870 G + 0.2989 R

‡∏à‡∏∞‡πÄ‡∏´‡πá‡∏ô‡∏ß‡πà‡∏≤ Green ‡∏°‡∏µ‡∏Ñ‡πà‡∏≤‡∏°‡∏≤‡∏Å‡∏™‡∏∏‡∏î‡πÅ‡∏õ‡∏•‡∏ß‡πà‡∏≤ -> **‡∏Ñ‡∏ß‡∏≤‡∏°‡πÄ‡∏Ç‡πâ‡∏°‡∏ô‡πâ‡∏≠‡∏¢** ‡πÄ‡∏û‡∏£‡∏≤‡∏∞‡∏ï‡∏≤‡∏Ñ‡∏ô‡πÄ‡∏£‡∏≤‡∏£‡∏±‡∏ö‡∏™‡∏µ‡πÄ‡∏Ç‡∏µ‡∏¢‡∏ß‡πÑ‡∏î‡πâ‡∏î‡∏µ‡∏Å‡∏ß‡πà‡∏≤‡πÅ‡∏î‡∏á, ‡∏ô‡πâ‡∏≥‡πÄ‡∏á‡∏¥‡∏ô

````python
img = cv2.imread('./week_introduction/lena.png', cv2.IMREAD_COLOR) 
row,col, n =img.shape
print(row, col, n)
````

````python
gray = np.zeros([row, col], dtype='float')# ‡∏™‡∏£‡πâ‡∏≤‡∏á array 10x10 ‡∏ó‡∏µ‡πà‡∏°‡∏µ‡∏Ñ‡πà‡∏≤‡πÄ‡∏õ‡πá‡∏ô 0 ‡πÅ‡∏•‡∏∞ type ‡πÄ‡∏õ‡πá‡∏ô float
print(gray)
````

### ‡∏ó‡∏≥‡πÑ‡∏°‡∏ï‡πâ‡∏≠‡∏á‡πÅ‡∏õ‡∏•‡∏á‡πÄ‡∏õ‡πá‡∏ô float16 ‡∏Å‡πà‡∏≠‡∏ô ‡πÅ‡∏•‡πâ‡∏ß‡∏Ñ‡πà‡∏≠‡∏¢‡πÅ‡∏õ‡∏•‡∏á‡∏Å‡∏•‡∏±‡∏ö‡πÄ‡∏û‡∏∑‡πà‡∏≠‡∏î‡∏π‡∏ú‡∏•‡∏•‡∏±‡∏û‡∏ò‡πå ?

‡πÄ‡∏û‡∏£‡∏≤‡∏∞‡∏£‡∏π‡∏õ‡∏ó‡∏µ‡πà‡πÑ‡∏î‡πâ‡∏à‡∏≤‡∏Å cv ‡∏à‡∏∞‡πÄ‡∏õ‡πá‡∏ô int ‡πÅ‡∏ï‡πà‡πÄ‡∏£‡∏≤‡∏à‡∏∞‡∏ó‡∏≥ garyscale ‡∏ã‡∏∂‡πà‡∏á‡∏ï‡πâ‡∏≠‡∏á‡∏Ñ‡∏π‡∏ì‡∏Ñ‡πà‡∏≤‡∏™‡∏±‡∏°‡∏õ‡∏£‡∏∞‡∏™‡∏¥‡∏ó‡∏ó‡∏µ‡πà‡πÄ‡∏õ‡πá‡∏ô‡∏ó‡∏®‡∏ô‡∏¥‡∏¢‡∏°

‡∏à‡∏∂‡∏á‡∏ï‡πâ‡∏≠‡∏á‡πÅ‡∏õ‡∏•‡∏á‡πÄ‡∏õ‡πá‡∏ô float16 -> ‡∏ó‡∏≥ garyscale -> ‡πÅ‡∏õ‡∏•‡∏á‡πÄ‡∏õ‡πá‡∏ô int ‡πÄ‡∏û‡∏∑‡πà‡∏≠‡∏î‡∏π‡∏ú‡∏•‡∏•‡∏±‡∏û‡∏ò‡πå

````python
img = img.astype(np.float16) #‡πÅ‡∏õ‡∏•‡∏á type ‡∏Ç‡∏≠‡∏á img ‡πÄ‡∏õ‡πá‡∏ô float16 ‡∏Å‡πà‡∏≠‡∏ô‡∏ó‡∏µ‡πà‡∏à‡∏∞ process ‡∏£‡∏π‡∏õ
for i in range(row):
    for j in range(col):
        gray[i, j] = 0.1140*img[i, j, 0] + 0.5870*img[i, j, 1] + 0.2989*img[i, j, 2]
````

````python
gray = gray.astype(np.uint8) # ‡πÅ‡∏õ‡∏•‡∏á type ‡∏Ç‡∏≠‡∏á gray ‡∏Å‡∏•‡∏±‡∏ö‡πÄ‡∏õ‡πá‡∏ô uint8‡πÄ‡∏û‡∏∑‡πà‡∏≠‡πÉ‡∏´‡πâ‡πÅ‡∏™‡∏î‡∏á‡∏ú‡∏•‡πÑ‡∏î‡πâ‡∏ñ‡∏π‡∏Å‡∏ï‡πâ‡∏≠‡∏á
````

````python
show_image(gray)
````
